```python
import seaborn as sns
import matplotlib.pyplot as plt
import math

def detect_and_plot_outliers(dataframe, numerical_columns):
    # Set up the subplots
    num_cols = len(numerical_columns)
    num_rows = math.ceil(num_cols / 3)
    
    plt.figure(figsize=(15, 5 * num_rows))
    
    for i, col in enumerate(numerical_columns, 1):
        plt.subplot(num_rows, 3, i)
        
        # Box plot
        sns.boxplot(x=dataframe[col], color='skyblue', width=0.5, linewidth=1.5, fliersize=5)
        plt.title(f'Box Plot of {col}')
        plt.xlabel(col)
        
        # Detect and mark outliers using the IQR method
        Q1 = dataframe[col].quantile(0.25)
        Q3 = dataframe[col].quantile(0.75)
        IQR = Q3 - Q1
        
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        outliers = (dataframe[col] < lower_bound) | (dataframe[col] > upper_bound)
        plt.scatter(dataframe[col][outliers], dataframe[col][outliers], color='red', marker='x', s=50, label='Outliers')
    
    plt.tight_layout()
    plt.show()
```

```python
detect_and_plot_outliers(hr, num_cols)
```

```python
# Filter object type columns without Attrition
obj_columns = hr.select_dtypes(include='object').drop(['Attrition'], axis=1)

# Define custom color palette
custom_palette = ['#1F77B4', '#AEC7E8']

# Plotting the distribution of each object column grouped by Attrition
for column in obj_columns.columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=hr, x=column, hue='Attrition', palette=custom_palette, edgecolor=".6", linewidth=1.5)
    plt.title(f"Distribution of {column} by Attrition")
    plt.xlabel(column)
    plt.ylabel("Frequency")
    plt.xticks(rotation=45)
    plt.legend(title='Attrition', loc='upper right')
    
    # Add annotations
    for p in plt.gca().patches:
        plt.gca().annotate(f"{p.get_height()}", (p.get_x() + p.get_width() / 2., p.get_height()),
                           ha='center', va='center', xytext=(0, 10), textcoords='offset points')
    
    plt.tight_layout()
    plt.show()
```

```python
def generate_advanced_pie_charts(dataframe, categorical_columns):
    for column in categorical_columns:
        plt.figure(figsize=(15, 15))
        
        # Count the occurrences of each unique value in the column
        category_counts = dataframe[column].value_counts()
        
        # Define a custom color palette
        custom_palette = sns.color_palette("Paired", len(category_counts))
        
        # Generate a pie chart using the custom color palette
        plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=90, colors=custom_palette)
        
        plt.title(f'Pie Chart of the column "{column}"')
        plt.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle.
        
        plt.show()
generate_advanced_pie_charts(hr, obj_cols)
```

```python
# Selecting only the numeric columns
numeric_columns = hr.select_dtypes(include='number')

# Calculate the correlation matrix
corr_matrix = numeric_columns.corr()

# Creating the heatmap using Seaborn and Matplotlib
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', cbar=True, square=True, linewidths=.5, annot_kws={'size': 10})
plt.title("Correlation Heatmap", fontsize=16)
plt.show()
```

```python
# Split the dataframe based on the Attrition column
attr_yes_df = hr[hr['Attrition'] == 'Yes']
attr_no_df = hr[hr['Attrition'] == 'No']

# Calculate value counts for EducationField and JobRole
dept_count_yes = attr_yes_df['EducationField'].value_counts()
dept_count_no = attr_no_df['EducationField'].value_counts()
job_role_yes = attr_yes_df['JobRole'].value_counts()
job_role_no = attr_no_df['JobRole'].value_counts()

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(20, 6))

# Pie chart for EducationField distribution
axes[0].pie(dept_count_yes * 100 / dept_count_no, wedgeprops=dict(width=0.6), labels=(dept_count_yes / dept_count_no).index, autopct='%1.1f%%', startangle=180)
axes[0].text(0., 0., '', horizontalalignment='center', verticalalignment='center')
axes[0].axis('equal')
axes[0].set_title('Education Field Distribution by Attrition')

# Pie chart for JobRole distribution
axes[1].pie(job_role_yes * 100 / job_role_no, wedgeprops=dict(width=0.6), labels=(job_role_yes / job_role_no).index, autopct='%1.1f%%', startangle=180)
axes[1].text(0., 0., '', horizontalalignment='center', verticalalignment='center')
axes[1].axis('equal')
axes[1].set_title('Job Role Distribution by Attrition')

plt.show()
```
